```python
dfn = pd.read_csv("/content/grades_withnulls.csv")
dfn
```

```python
mv = dfn['1stsem'].mean()
dfn = dfn.fillna(mv)
print(dfn)
```

```python
dfn = dfn.drop_duplicates()
print(dfn)
```

```python
dfn.rename(columns = {'Grade':'GPA'}, inplace = True)
dfn.head()
```

```python
dfn.columns = ['Student_name', 'Mr/Ms', '1stsem', '2ndsem','3rdsem',
'F_Outcome','Placed?']
dfn.head()
```

```python
dfn['Avg_score'] = (dfn['1stsem']+dfn['2ndsem']+dfn['3rdsem'])/3
dfn.head()
```

```python
dfn.groupby('Placed?')['Avg_score'].agg(['min', 'max'])
```
