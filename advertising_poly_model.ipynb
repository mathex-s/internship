```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/content/Advertising.csv")
print(df)
```

```python
print(df.isnull().sum())
```

```python
print(df.head())
```

```python
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)
print(X_poly)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)
print(y_train)
```

```python
PR = LinearRegression()
PR.fit(X_train, y_train)
```

```python
TV = float(input("Enter TV advertising spend ($): "))
Radio = float(input("Enter Radio advertising spend ($): "))
Newspaper = float(input("Enter Newspaper advertising spend ($): "))
new_data = np.array([[TV, Radio, Newspaper]])
new_data_poly = poly.transform(new_data)
predicted_sales = PR.predict(new_data_poly)[0]
print(f"\nPredicted Sales: {predicted_sales:.2f} units")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))

# Define tv_range for plotting the regression line
tv_range = np.linspace(df['TV'].min(), df['TV'].max(), 100).reshape(-1, 1)
radio_mean = df['Radio'].mean()
newspaper_mean = df['Newspaper'].mean()

# Create a DataFrame for prediction with constant Radio and Newspaper
# Need to construct the features in the same order as X for poly.transform
# The original X was ['TV', 'Radio', 'Newspaper']
plot_data = pd.DataFrame({
    'TV': tv_range.flatten(),
    'Radio': np.full_like(tv_range.flatten(), radio_mean),
    'Newspaper': np.full_like(tv_range.flatten(), newspaper_mean)
})

# Transform the plotting data using the fitted PolynomialFeatures
plot_data_poly = poly.transform(plot_data)

# Predict sales for the plotting data
predicted_sales_for_plot = PR.predict(plot_data_poly)

# Scatter plot of actual sales
sns.scatterplot(x=df['TV'], y=df['Sales'], label='Actual Sales', alpha=0.6)

# Line plot of predicted sales
sns.lineplot(x=tv_range.flatten(), y=predicted_sales_for_plot, color='red', label='Predicted Sales')

plt.title('Polynomial Regression: Sales vs. TV Advertising (Radio & Newspaper at Mean)')
plt.xlabel('TV Advertising Spend ($)')
plt.ylabel('Sales (Units)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
y_pred = PR.predict(X_test)
# Calculate evaluation metrics
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f"R-squared: {r2:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```
