```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
Img_size=224
batch_size=32
```

```python
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    validation_split=0.2
)
```

```python
train_generator=train_datagen.flow_from_directory(
    "/content/drive/MyDrive/brain_tumor_dataset/train",
    target_size=(Img_size,Img_size),
    batch_size=batch_size,
    class_mode='binary',
    subset='training'
)
```

```python
val_generator=train_datagen.flow_from_directory(
    "/content/drive/MyDrive/brain_tumor_dataset/train",
    target_size=(Img_size,Img_size),
    batch_size=batch_size,
    class_mode='binary',
    subset='training'
)
```

```python
model=keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(Img_size, Img_size, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid'),
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_generator,epochs=5,validation_data=val_generator,batch_size=batch_size)
```

```python
model.save("/content/drive/MyDrive/brain_tumor_dataset/train/braintumer.h5")
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
model=load_model("/content/drive/MyDrive/brain_tumor_dataset/train/braintumer.h5")
print("model loaded")
```

```python
test_image_path="/content/drive/MyDrive/brain_tumor_dataset/train/yes/Y1721.jpg"
img=image.load_img(test_image_path,target_size=(224,224))
plt.imshow(img)
plt.axis()
plt.show()
```

```python
img_array=image.img_to_array(img)
img_array=np.expand_dims(img_array,axis=0)
img_array=img_array/255.0
```

```python
prediction=model.predict(img_array)
print(prediction)
```

```python
if prediction > 0.5:
  print("tumor detected")
else:
  print("no tumor detected")
```
